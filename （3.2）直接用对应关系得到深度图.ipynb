```python
import numpy as np
import matplotlib.pyplot as plt
import time
import cv2
%matplotlib inline
```

```python
path = 'photo\p0\\'

corresponding_in = np.load(path+'small_corresponding_in.npy')
# corresponding_in = np.load(path+'tps_small_corresponding_in.npy')

lis =np.load(path+'small_corresponding_contours.npy')
data1 = np.load(path+'small_contours.npy')
data2 = np.load(path+'small_contours_smpl.npy')
```

```python

depthMap =cv2.imread(path+'depthMap.png',0)
```

```python
# 直接将深度图的像素值作为深度值
newPng_origin = np.mat(np.zeros((depthMap.shape)))
for i in corresponding_in:
    newPng_origin[i[0][1],i[0][0]] = depthMap[i[1][1], i[1][0]]
    
for i in lis:
#     print(data1[int(i[0])][1],data1[int(i[0])][0])
    newPng_origin[data1[int(i[0])][1],data1[int(i[0])][0]] = depthMap[data2[int(i[1])][1], data2[int(i[1])][0]]
 
newPng_origin_array = newPng_origin.A
cv2.imwrite(path+"newPng_origin.png", newPng_origin_array)
# cv2.imwrite(path+"tps_newPng_origin.png", newPng_origin_array)
```
